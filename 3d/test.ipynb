```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 3D Scatter Plot Data
x_scatter = np.random.rand(100) * 10  # 100 random points between 0 and 10
y_scatter = np.random.rand(100) * 10
z_scatter = x_scatter + y_scatter  # z = x + y
```

```python
# Create a figure for the scatter plot
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter3D(x_scatter, y_scatter, z_scatter, c='r', marker='o')
ax1.set_title('3D Scatter Plot')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_zlabel('Z Axis')
plt.show()
```

```python
# 3D Surface Plot Data
x_surface = np.linspace(-5, 5, 100)
y_surface = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_surface, y_surface)
Z = np.sin(np.sqrt(X**2 + Y**2))  # z = sin(sqrt(x^2 + y^2))
```

```python
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_title('3D Surface Plot')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.set_zlabel('Z Axis')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot Data
x_scatter = np.random.rand(100) * 10  # 100 random points between 0 and 10
y_scatter = np.random.rand(100) * 10
z_scatter = x_scatter + y_scatter  # z = x + y

# Create a figure for the scatter plot
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter3D(x_scatter, y_scatter, z_scatter, c='r', marker='o')
ax1.set_title('3D Scatter Plot')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_zlabel('Z Axis')

# 3D Surface Plot Data
x_surface = np.linspace(-5, 5, 100)
y_surface = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_surface, y_surface)
Z = np.sin(np.sqrt(X**2 + Y**2))  # z = sin(sqrt(x^2 + y^2))

# Create a figure for the surface plot
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_title('3D Surface Plot')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.set_zlabel('Z Axis')

# Show the plots
plt.tight_layout()
plt.show()
```
